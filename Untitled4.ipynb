```python
def exibir_tabuleiro(tabuleiro):
    """Exibe o tabuleiro do Jogo da Velha."""
    print(f"\n {tabuleiro[0]} | {tabuleiro[1]} | {tabuleiro[2]} ")
    print("-----------")
    print(f" {tabuleiro[3]} | {tabuleiro[4]} | {tabuleiro[5]} ")
    print("-----------")
    print(f" {tabuleiro[6]} | {tabuleiro[7]} | {tabuleiro[8]} \n")

def verificar_vitoria(tabuleiro, jogador):
    """Verifica se o jogador atual venceu."""
    # Todas as combina√ß√µes vencedoras (linhas, colunas, diagonais)
    combinacoes_vencedoras = [
        # Linhas
        [0, 1, 2], [3, 4, 5], [6, 7, 8],
        # Colunas
        [0, 3, 6], [1, 4, 7], [2, 5, 8],
        # Diagonais
        [0, 4, 8], [2, 4, 6]
    ]

    for combo in combinacoes_vencedoras:
        if tabuleiro[combo[0]] == jogador and \
           tabuleiro[combo[1]] == jogador and \
           tabuleiro[combo[2]] == jogador:
            return True
    return False

def verificar_empate(tabuleiro):
    """Verifica se o jogo terminou em empate."""
    return ' ' not in tabuleiro # Retorna True se n√£o houver mais espa√ßos vazios

def jogar_jogo_da_velha():
    """Fun√ß√£o principal que executa o jogo."""
    while True:
        tabuleiro = [' '] * 9 # Inicializa o tabuleiro com 9 espa√ßos vazios
        jogador_atual = 'X'

        print("Bem-vindo ao Jogo da Velha!")
        print("Use os n√∫meros de 1 a 9 para escolher a posi√ß√£o:")
        exibir_tabuleiro(['1', '2', '3', '4', '5', '6', '7', '8', '9'])

        jogo_ativo = True
        while jogo_ativo:
            exibir_tabuleiro(tabuleiro)

            # Pede a jogada do jogador atual
            while True:
                try:
                    posicao = int(input(f"Jogador {jogador_atual}, sua vez (1-9): ")) - 1
                    if 0 <= posicao <= 8 and tabuleiro[posicao] == ' ':
                        tabuleiro[posicao] = jogador_atual
                        break # Sai do loop de valida√ß√£o se a jogada for v√°lida
                    else:
                        print("Posi√ß√£o inv√°lida ou j√° ocupada. Tente novamente.")
                except ValueError:
                    print("Entrada inv√°lida. Digite um n√∫mero.")

            # Verifica vit√≥ria ou empate
            if verificar_vitoria(tabuleiro, jogador_atual):
                exibir_tabuleiro(tabuleiro)
                print(f"Parab√©ns! O Jogador {jogador_atual} venceu!")
                jogo_ativo = False
            elif verificar_empate(tabuleiro):
                exibir_tabuleiro(tabuleiro)
                print("O jogo terminou em empate!")
                jogo_ativo = False
            else:
                # Troca o jogador
                jogador_atual = 'O' if jogador_atual == 'X' else 'X'

        # Pergunta se quer jogar de novo
        if input("Jogar novamente? (s/n): ").lower() != 's':
            print("Obrigado por jogar!")
            break

# Inicia o jogo
if __name__ == "__main__":
    jogar_jogo_da_velha()
```
